```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
r = requests.get('https://butterfly-conservation.org/uk-butterflies/a-to-z', verify=False)
```

```python
soup = BeautifulSoup(r.text)
```

```python
links = soup.find_all('a')
links
```

```python
for link in links:
    print(link.attrs.get('href'))
```

```python
hrefs = [link.attrs.get('href') for link in links]
```

```python
butterfly_pages = hrefs[39:100]
```

```python
urls = ['https://butterfly-conservation.org' + page for page in butterfly_pages]
urls
```

```python
def peel_data_from_element(element):
    just_text = element.text
    return just_text.split(': ')[1]
```

```python
def get_butterfly(url):
    
    r = requests.get(url, verify=False)
    soup = BeautifulSoup(r.text)
    
    h1 = soup.find('h1')
    
    name = h1.text
    name = name.strip()
    
    family = soup.find('li', text=re.compile(r'Family:*'))
    family_data = peel_data_from_element(family)
    
    size = soup.find('li', text=re.compile(r'Size:*'))
    size_data = peel_data_from_element(size)
    
    wing_span = soup.find('li', text=re.compile(r'Wing Span:*'))
    wing_span_data = peel_data_from_element(wing_span)
    
    return{'name': name, 'family': family_data, 'size': size_data, 'wing span': wing_span_data}
```

```python
data = get_butterfly("https://butterfly-conservation.org/butterflies/northern-brown-argus")
```

```python
def process_each_link(list):
    data = []
    for url in list:
        try:
            data.append(get_butterfly(url))
        except:
            print('Missed a butterfly at ', url)
    return data
```

```python
test_list = ['https://butterfly-conservation.org/butterflies/chalk-hill-blue',
             'https://butterfly-conservation.org/butterflies/chequered-skipper',
             'https://butterfly-conservation.org/butterflies/clouded-yellow']
```

```python
butterfly_data = [process_each_link(test_list)]
```

```python
print(butterfly_data)
```

```python
butterfly_data = process_each_link(urls)
```

```python
import csv
```

```python
def write_csv(data_list):
    with open('butterflies.csv', 'w', newline='') as f:
        fieldnames = ['name', 'family', 'size', 'wing span']
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data_list)
```

```python
write_csv(butterfly_data)
```

```python
with open('butterflies.csv', newline='') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```
